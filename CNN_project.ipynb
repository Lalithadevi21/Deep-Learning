#Model Building

```python
import tensorflow as tf
```

```python
from tensorflow.keras.models import Sequential
```

```python
from tensorflow.keras.layers import Input, Conv2D, MaxPooling2D, Flatten, Dense
```

```python
X_train.shape
```

```python
Cnn_model = Sequential()
```

```python
Cnn_model.add(Input(shape = (48,48,3)))
```

#First Convolutional Layer

```python
Cnn_model.add(Conv2D(filters = 32, kernel_size = (3,3),strides =(1,1),padding ="valid", activation = "relu"))
```

```python
Cnn_model.add(MaxPooling2D(pool_size=(2,2), padding = "valid"))
```

#Second Convolutional Layer

```python
Cnn_model.add(Conv2D(filters = 32, kernel_size = (3,3),strides =(1,1),padding ="valid", activation = "relu"))
```

```python
Cnn_model.add(MaxPooling2D(pool_size=(2,2), padding = "valid"))
```

#Third Convolutional Layer

```python
Cnn_model.add(Conv2D(filters = 32, kernel_size = (3,3),strides =(1,1),padding ="valid", activation = "relu"))
```

```python
Cnn_model.add(MaxPooling2D(pool_size=(2,2), padding = "valid"))
```

#Fully connected Feed Forward Neural Network

```python
Cnn_model.add(Flatten())
```

```python
Cnn_model.add(Dense(256,activation = "relu"))
```

```python
Cnn_model.add(Dense(512,activation = "relu"))
```

```python
Cnn_model.add(Dense(1024,activation = "relu"))
```

```python
Cnn_model.add(Dense(512,activation = "relu"))
```

```python
Cnn_model.add(Dense(256,activation = "relu"))
```

```python
Cnn_model.add(Dense(2,activation = "softmax"))
```

#Summary

```python
Cnn_model.summary()
```

#Compile

```python
Cnn_model.compile(optimizer = "adam",
                  loss = "sparse_categorical_crossentropy",
                  metrics =["accuracy"])
```

```python
Cnn_model.fit(X_train,y_train,epochs=15)
```

```python
Cnn_model.evaluate(X_test,y_test)
```
